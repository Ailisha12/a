```python
from vosk import Model, KaldiRecognizer
import sys
import os
import wave
import json

# if not os.path.exists("model-en"):
#     print ("Please download the model from https://github.com/alphacep/kaldi-android-demo/releases and unpack as 'model-en' in the current folder.")
#     exit (1)

wf = wave.open("/content/drive/MyDrive/Machine Learning ITHUB/ДИ 2023/lessons/data/KHleb_-_CHajj_sakhar_62509073 (1).wav", "rb")
if wf.getnchannels() != 1 or wf.getsampwidth() != 2 or wf.getcomptype() != "NONE":
    print ("Audio file must be WAV format mono PCM.")
    exit (1)

model = Model("/content/vosk-model-small-ru-0.22")
rec = KaldiRecognizer(model, wf.getframerate())

result = ""
while True:
    data = wf.readframes(10000)
    if len(data) == 0:
        break
    if rec.AcceptWaveform(data):
      r = rec.Result()
      r = json.loads(r)
      if len(r['text'])!=0:
        print(r['text'])
      # result+=r['text']
      # result+=" "
    else:
      pr = rec.PartialResult()
      pr = json.loads(pr)
      if len(pr['partial'])!=0:
        print(pr['partial'])
      # result+=pr['partial']
      # result+=" "
```
