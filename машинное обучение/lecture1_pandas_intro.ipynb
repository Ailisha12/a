```python
#задача - оставить в датафрейме фильмы топ-5 популярных жанров и вывести топ-20 фильмов по сборам(loc)
```

```python
top_genres=df['genre1'].value_counts().head()
top_genres=list(top_genres.index)
```

```python
df_genres=df.loc[df['genre1'].isin(top_genres)]
df_genres.sort_values(by='revenue',ascending=False).head(20)
```
