```python
#импорт библиотек
import pandas as pd
import numpy as np
```

```python
#считаем файл с екселем в датафрейм
films=pd.read_excel('data\moviedb.xlsx')
films.head()
```

```python
#решение задачи #4 (lecture1)
#задача - создать новый столбец profit в датафрейме (разность сборов и бюджета). Вывести топ-10 фильмов по окупаемости(profit)
films['budget']=films['budget']//1000000
films['revenue']=films['revenue']//1000000
films.head()
```

```python
films['profit']=films['revenue']-films['budget']
films.head()
```

```python
films.sort_values(by='profit',ascending=False).head(10)
```

```python
#удаление столбцов (drop, inplace)
#axis=1 - столбцы
#axis=0 - строки
films.drop(['profit'],axis=1,inplace=True)
```

```python
films.head()
```

```python
films.drop([films.columns[0]],axis=1,inplace=True)
films.head()
```

```python
films_1 = films[['title','genre1','rating']]
films_2 = films[['budget','revenue']]
films_1.head()
```

```python
#объединение столбцов - concat по осям
films_merged = pd.concat([films_1,films_2],axis=1)
films_merged.head()
```

```python
films_merged.info()
```

```python
#работа с пропусками - dropna(subset), fillna(value)
films_drop=films_merged.dropna()
films_drop.info()
```

```python
films_drop = films_merged.dropna(subset=['budget'])
films_drop.info()
```

```python
#группировка
top_countries=films.groupby(by='country1').agg(
    {
        'rating':['mean','count']
    }
)
top_countries.columns
top_countries=top_countries[top_countries[('rating', 'count')]>100]
top_countries.sort_values(by=('rating',  'mean'),ascending=False)
```

```python
#сводные таблицы
pd.pivot_table(films,index=['genre1'],columns=['title'],values='rating',aggfunc=np.mean)
```
